```python
# import modules
import os
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# files to load
file1 = "ca_ev_registrations.csv"
file2 = "ca_emissions.csv"
file3 = "ca_energy.csv"

# read files and store into pandas dataframe
registrations = pd.read_csv(file1, low_memory=False)
emissions = pd.read_csv(file2, low_memory=False)
energy = pd.read_csv(file3, low_memory=False)

registrations.head()
```

```python
# remove DMV ID, DMV snapshot, and registration expiration date
registrations = registrations.drop(columns=['DMV ID', 'DMV Snapshot', 'Registration Expiration Date'])
registrations.head()
```

```python
# diplay the CA emissions dataframe
emissions.head()
```

```python
# remove empty rows 0 through 2 from dataframe
emissions = emissions.drop(labels=range(0, 3), axis=0)
emissions.head()
```

```python
# remove empty rows 10 through 20 from dataframe
emissions = emissions.drop(labels=range(10, 21), axis=0)
emissions.head()
```
