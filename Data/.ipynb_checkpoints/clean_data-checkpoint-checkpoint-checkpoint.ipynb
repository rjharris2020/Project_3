```python
# merge all top 5 dataframes for each year into one dataframe
top5_ev = year_2010.append(year_2011).append(year_2012).append(year_2013).append(year_2014).append(year_2015).append(year_2016).append(year_2017).append(year_2018).append(year_2019).append(year_2020)
top5_ev
```

```python
# create 'Base Model' column based on make of EV
top5_ev['Base Model'] = top5_ev['Vehicle Name'].str.split(" ").str[0]
```

```python
#display the dataframe
top5_ev
```

```python
# set new variable for ev consumption dataframe
new_evconsumption = ev_consumptiondf
```

```python
# create 'Base Model' column based on make of EV
new_evconsumption['Base Model'] = new_evconsumption['Vehicle Name'].str.split(" ").str[0]
```

```python
# display dataframe
new_evconsumption
```

```python
# drop all duplicates from evconsumption dataframe
new_evconsumption = new_evconsumption.drop_duplicates(subset=['Base Model'], keep='first')
```

```python
new_evconsumption
```

```python
# drop 'Vehicle Name' column and display dataframe
new_evconsumption = new_evconsumption.drop(columns='Vehicle Name')
new_evconsumption
```

```python
# merge the top 5 ev dataframe with the ev consumption
final_top5 = top5_ev.merge(new_evconsumption, how='left', on='Base Model')
final_top5
```

```python
# check for null values
final_top5[final_top5['EV Consumption(195 Wh/km)'].isnull()]
```

```python
# enter the EV consumption of 210 Wh/km for Chevy Volts
final_top5['EV Consumption(195 Wh/km)'] = final_top5['EV Consumption(195 Wh/km)'].fillna(210)
```

```python
final_top5
```

```python
# add carbon emissions column based on the average miles driven of 14,263 miles
final_top5['Carbon Emission'] = 14263 * 1.609  * (final_top5['EV Consumption(195 Wh/km)'].astype(int) / 1000000) * 496.54
final_top5
```

```python
final_top5 = final_top5.set_index('Year')
final_top5
```

```python
# create new dataframe into csv
final_ev = final_top5.to_csv('final_top5ev.csv', index = True, header = True)
```
